```python
def data_locator(exp_tag, problem_name, optimizer_name,num_arms, num_dim):
    return f"/home/juju/projects/bbo/results/{exp_tag}/{num_arms}arm/{num_dim}d/{problem_name}/1iter/{optimizer_name}"
def get_perf(mu_sobol, mu_mtv, sg_sobol, sg_mtv, exp_name, optimizers, num_arms, num_dim, i_agg=-1):
    mu, sg = ap.plot_agg_dim_all(data_locator, exp_name, num_arms, num_dim, optimizers=optimizers, sort=True, i_agg=i_agg)
    mu_mtv.append(mu[0])
    mu_sobol.append(mu[1])
    sg_mtv.append(sg[0])
    sg_sobol.append(sg[1])
    return mu_sobol, mu_mtv, sg_sobol, sg_mtv

def get_compare(exp_name, num_arms, num_dim, i_agg=-1):
    optimizers = [
        f"sobol_{num_arms}arm",
        f'mtv_{num_arms}arm',
    ]
    mu_sobol =[]
    mu_mtv = []
    sg_sobol=[]
    sg_mtv = []
    for i in range(len(num_dim)):
        mu_sobol, mu_mtv, sg_sobol, sg_mtv=get_perf(mu_sobol, mu_mtv, sg_sobol, sg_mtv, exp_name, optimizers, num_arms, num_dim[i], i_agg)
    plt.errorbar(num_dims, mu_sobol, sg_sobol, fmt='+', label='sobol', linestyle='None', color = 'grey')
    plt.errorbar(num_dims, mu_mtv, sg_mtv, fmt='o', label='mtv', linestyle='None', color = 'black')
    plt.title(f"compare mtv and sobol in different dimensions with {num_arms}arms")
    plt.legend()
    return mu_sobol, mu_mtv, sg_sobol, sg_mtv
```
