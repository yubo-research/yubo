# Ackley

```python
def ackley(x, y):
    a = 20
    b = 0.2
    c = 2 * np.pi
    term1 = -a * np.exp(-b * np.sqrt((x**2 + y**2) / 2))
    term2 = -np.exp((np.cos(c * x) + np.cos(c * y)) / 2)
    return term1 + term2 + a + np.exp(1)

# Create a grid of points
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = ackley(X, Y)
name = "Ackley"
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.contour3D(X, Y, Z, 50, cmap='gray')

# ax.plot_surface(X, Y, Z, cmap='gray',  rstride=10, cstride=5, alpha=1)
# ax.plot_wireframe(X, Y, Z, color="black")
# ax.set_proj_type('persp', focal_length=0.3)
# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel(f'{name} Function')
ax.view_init(20, 30)
# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# # make the grid lines transparent
# ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
plt.savefig(f'figures/bench/{name}.png', bbox_inches='tight', pad_inches = 0)
# Show the plot
# plt.title('3D Plot of the Ackley Function')
plt.show()
```

# Branin

```python
def branin(x, y):
    a = 1
    b = 5.1 / (4 * np.pi ** 2)
    c = 5 / np.pi
    r = 6
    s = 10
    t = 1 / (8 * np.pi)

    term1 = a * (y - b * x ** 2 + c * x - r) ** 2
    term2 = s * (1 - t) * np.cos(x)

    return term1 + term2 + s

# Create a grid of points
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = branin(X, Y)
name = "Branin"
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
# ax.contour3D(X, Y, Z, 50, cmap='cool')
# ax.plot_surface(X, Y, Z, cmap='gray')
ax.plot_surface(X, Y, Z, cmap='gray',  rstride=10, cstride=5, alpha=1)
# ax.plot_wireframe(X, Y, Z, color="black")
# ax.set_proj_type('persp', focal_length=0.3)
# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
# ax.set_zlabel(f'{name} Function')
ax.view_init(20, 30)
# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# # make the grid lines transparent
# ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
plt.savefig(f'figures/bench/{name}.png', bbox_inches='tight')
# Show the plot
# plt.title('3D Plot of the Ackley Function')
plt.show()
```

# Drop Wave

```python
def drop_wave(x, y):
    numerator = 1 + np.cos(12 * np.sqrt(x**2 + y**2))
    denominator = 0.5 * (x**2 + y**2) + 2
    return -numerator / denominator
# Create a grid of points
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = drop_wave(X, Y)
name = "DropWave"
# Create a 3D plot
fig = plt.figure(figsize=(7, 20))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
# ax.contour3D(X, Y, Z, 50, cmap='gray')
ax.plot_surface(X, Y, Z, cmap='gray')
# ax.plot_surface(X, Y, Z, cmap='gray',  rstride=10, cstride=5, alpha=1)
# ax.plot_wireframe(X, Y, Z, color="black")
# ax.set_proj_type('persp', focal_length=0.3)
# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
# ax.set_zlabel(f'{name} Function')
ax.view_init(25, 25)
# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# # make the grid lines transparent
# ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
plt.savefig(f'figures/bench/{name}.png', bbox_inches='tight')
# Show the plot
# plt.title('3D Plot of the Ackley Function')
plt.show()
```

# DixonPrice

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the Dixon-Price function
def dixon_price(x):
    n = len(x)
    result = (x[0] - 1)**2
    for i in range(1, n):
        result += (i + 1) * (2 * x[i]**2 - x[i - 1])**2
    return result
name = "DixonPrice"
# Create a grid of points for x and y
# x_range = np.linspace(-10, 10, 400)
# y_range = np.linspace(-10, 10, 400)
x_range = np.linspace(-5, 5, 500)
y_range = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x_range, y_range)
Z = np.zeros_like(X)

# Evaluate the Dixon-Price function for the grid of points
for i in range(len(x_range)):
    for j in range(len(y_range)):
        Z[i, j] = dixon_price([X[i, j], Y[i, j]])

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='gray')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.view_init(25, 25)
# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# ax.set_zlabel('f(X, Y) - Dixon-Price')
# ax.set_title('Dixon-Price Function')
plt.savefig(f'figures/bench/{name}.png', bbox_inches='tight')

# Show the plot
plt.show()
```

# Eggholder

```python
def eggholder(x, y):
    return (-(y + 47) * np.sin(np.sqrt(np.abs(x / 2 + (y + 47)))) - x * np.sin(np.sqrt(np.abs(x - (y + 47)))))
```

```python
plotfig(eggholder, "EggHolder")
```

# Griewant

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the Griewank function
def griewank(x, y):
    return (x**2 + y**2) / 4000 - np.cos(x) * np.cos(y / np.sqrt(2)) + 1
plotfig(griewank, "GrieWank")
```

# GrLee12

```python
def gramacy_lee(x, y):
    term1 = np.sin(10 * np.pi * x) / (2 * x)
    term2 = np.sin(10 * np.pi * y) / (2 * y)
    term3 = (x - 1) ** 4
    term4 = (y - 1) ** 4
    return term1 + term2 + term3 + term4
plotfig(gramacy_lee, "GrLee12")
```

# Holder Table 

```python
def holder_table(x, y):
    term1 = -np.abs(1 - np.sqrt(x**2 + y**2) / np.pi)
    term2 = np.sin(x) * np.cos(y) * np.exp(np.abs(1 - np.sqrt(x**2 + y**2) / np.pi))
    return term1 * term2
```

```python
plotfig(holder_table, "Holder Table")
```

# Levy 

```python
def levy(x, y):
    term1 = (np.sin(3 * np.pi * x))**2
    term2 = (x - 1)**2 * (1 + (np.sin(3 * np.pi * y))**2)
    term3 = (y - 1)**2 * (1 + (np.sin(2 * np.pi * y))**2)
    return term1 + term2 + term3
```

```python
plotfig(levy, "Levy")
```

```python
def michalewicz(x, y, m=10):
    term1 = np.sin(x) * (np.sin((x**2) / np.pi))**(2*m)
    term2 = np.sin(y) * (np.sin((2*y**2) / np.pi))**(2*m)
    return -term1 - term2
```

```python
plotfig(michalewicz, "Michalewicz")
```

# Powell

```python
def powell(x, y):
    term1 = (x + 10 * y)**2
    term2 = 5 * (x - y)**2
    term3 = (x - 2*y)**4
    term4 = 10 * (x - y)**4
    return term1 + term2 + term3 + term4
```

```python
plotfig(powell, "Powell")
```

# Rastrigin

```python
def rastrigin(x, y):
    A = 10
    return A * 2 + (x**2 - A * np.cos(2 * np.pi * x)) + (y**2 - A * np.cos(2 * np.pi * y))
```

```python
plotfig(rastrigin, "Rastrigin")
```

# Rosenbrock

```python
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2
```

```python
plotfig(rosenbrock, "Rosenbrock")
```

# Shubert

```python
def shubert(x1,x2):
      sum1=0
      sum2=0
      for i in range(1,6):
          sum1 = sum1 + (i* cos(((i+1)*x1) +i))
          sum2 = sum2 + (i* cos(((i+1)*x2) +i))
      return sum1 * sum2
```

```python
plotfig(shubert, "Shubert")
```

# SixHumpCamel

```python
def six_hump_camel(x, y):
    term1 = (4 - 2.1 * x**2 + (x**4) / 3) * x**2
    term2 = x * y
    term3 = (-4 + 4 * y**2) * y**2
    return term1 + term2 + term3
```

```python
plotfig(six_hump_camel, "SixHumpCamel")
```

# StybTang

```python
def styblinski_tang(x, y):
    return 0.5 * (x**4 - 16 * x**2 + 5 * x + y**4 - 16 * y**2 + 5 * y)
```

# Three Hump Camel

```python
def three_hump_camel(x0, x1):
    return 2.0 * x0**2 - 1.05 * x0**4 + x0**6 / 6.0 + x0 * x1 + x1**2
```

```python
plotfig(three_hump_camel,"ThreeHumpCamel")
```
