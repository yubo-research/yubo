# Ackley

```python
def ackley(x, y):
    a = 20
    b = 0.2
    c = 2 * np.pi
    term1 = -a * np.exp(-b * np.sqrt((x**2 + y**2) / 2))
    term2 = -np.exp((np.cos(c * x) + np.cos(c * y)) / 2)
    return term1 + term2 + a + np.exp(1)

# Create a grid of points
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = ackley(X, Y)
name = "Ackley"
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.contour3D(X, Y, Z, 50, cmap='gray')

# ax.plot_surface(X, Y, Z, cmap='gray',  rstride=10, cstride=5, alpha=1)
# ax.plot_wireframe(X, Y, Z, color="black")
# ax.set_proj_type('persp', focal_length=0.3)
# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel(f'{name} Function')
ax.view_init(20, 30)
# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# # make the grid lines transparent
# ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
plt.savefig(f'figures/bench/{name}.png', bbox_inches='tight')
# Show the plot
# plt.title('3D Plot of the Ackley Function')
plt.show()
```

# Beale

```python
def beale(x, y):
    term1 = (1.5 - x + x * y)**2
    term2 = (2.25 - x + x * y**2)**2
    term3 = (2.625 - x + x * y**3)**2
    return term1 + term2 + term3


# Create a grid of points
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = beale(X, Y)
name = "Beale"
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
# ax.contour3D(X, Y, Z, 50, cmap='cool')
ax.plot_surface(X, Y, Z, cmap='gray')
# ax.plot_surface(X, Y, Z, cmap='gray',  rstride=10, cstride=5, alpha=1)
# ax.plot_wireframe(X, Y, Z, color="black")
# ax.set_proj_type('persp', focal_length=0.3)
# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
# ax.set_zlabel(f'{name} Function')
ax.view_init(20, 30)
# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# # make the grid lines transparent
# ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
plt.savefig(f'figures/bench/{name}.png', bbox_inches='tight')
# Show the plot
# plt.title('3D Plot of the Ackley Function')
plt.show()
```

# Branin

```python
def branin(x, y):
    a = 1
    b = 5.1 / (4 * np.pi ** 2)
    c = 5 / np.pi
    r = 6
    s = 10
    t = 1 / (8 * np.pi)

    term1 = a * (y - b * x ** 2 + c * x - r) ** 2
    term2 = s * (1 - t) * np.cos(x)

    return term1 + term2 + s

# Create a grid of points
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = branin(X, Y)
name = "Branin"
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
# ax.contour3D(X, Y, Z, 50, cmap='cool')
# ax.plot_surface(X, Y, Z, cmap='gray')
ax.plot_surface(X, Y, Z, cmap='gray',  rstride=10, cstride=5, alpha=1)
# ax.plot_wireframe(X, Y, Z, color="black")
# ax.set_proj_type('persp', focal_length=0.3)
# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
# ax.set_zlabel(f'{name} Function')
ax.view_init(20, 30)
# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# # make the grid lines transparent
# ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
plt.savefig(f'figures/bench/{name}.png', bbox_inches='tight')
# Show the plot
# plt.title('3D Plot of the Ackley Function')
plt.show()
```

# CrossInTray

# Drop Wave

```python
def drop_wave(x, y):
    numerator = 1 + np.cos(12 * np.sqrt(x**2 + y**2))
    denominator = 0.5 * (x**2 + y**2) + 2
    return -numerator / denominator
# Create a grid of points
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = drop_wave(X, Y)
name = "DropWave"
# Create a 3D plot
fig = plt.figure(figsize=(7, 20))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
# ax.contour3D(X, Y, Z, 50, cmap='gray')
ax.plot_surface(X, Y, Z, cmap='gray')
# ax.plot_surface(X, Y, Z, cmap='gray',  rstride=10, cstride=5, alpha=1)
# ax.plot_wireframe(X, Y, Z, color="black")
# ax.set_proj_type('persp', focal_length=0.3)
# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
# ax.set_zlabel(f'{name} Function')
ax.view_init(25, 25)
# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# # make the grid lines transparent
# ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
plt.savefig(f'figures/bench/{name}.png', bbox_inches='tight')
# Show the plot
# plt.title('3D Plot of the Ackley Function')
plt.show()
```

# DixonPrice

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the Dixon-Price function
def dixon_price(x):
    n = len(x)
    result = (x[0] - 1)**2
    for i in range(1, n):
        result += (i + 1) * (2 * x[i]**2 - x[i - 1])**2
    return result
name = "DixonPrice"
# Create a grid of points for x and y
x_range = np.linspace(-10, 10, 400)
y_range = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x_range, y_range)
Z = np.zeros_like(X)

# Evaluate the Dixon-Price function for the grid of points
for i in range(len(x_range)):
    for j in range(len(y_range)):
        Z[i, j] = dixon_price([X[i, j], Y[i, j]])

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='gray')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
# ax.set_zlabel('f(X, Y) - Dixon-Price')
# ax.set_title('Dixon-Price Function')
plt.savefig(f'figures/bench/{name}.png', bbox_inches='tight')

# Show the plot
plt.show()
```
